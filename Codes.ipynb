```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.ones((500, 500, 3), dtype=np.uint8) * 255

count = 45
while count > 0:
    cv2.ellipse(img=img, center=(340, 250), axes=(80-count, 80-count), angle=-60, startAngle=0, endAngle=300, color=(10, 142, 255), thickness= 3)  
    count -= 1

count = 45
while count > 0:
    cv2.ellipse(img=img, center=(160, 250), axes=(80-count, 80-count), angle=0, startAngle=0, endAngle=300, color=(140, 219, 103), thickness= 3)  
    count -= 1

count = 45
while count > 0:
    cv2.ellipse(img=img, center=(250, 100), axes=(80-count, 80-count), angle=120, startAngle=0, endAngle=300, color=(255, 38, 66), thickness= 32) 
    count -= 1


cv2.putText(img=img, text='OpenCV', org=(80, 430), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=3, color=(0, 0, 0), thickness=12, lineType=cv2.LINE_AA)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.imshow(cv2.cvtColor(img_rgb , cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
